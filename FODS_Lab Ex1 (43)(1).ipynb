```python
import pandas as pd
```

```python
df=pd.DataFrame([{'Name':'Alice','Age':23,'Gender':'F','Score':85},{'Name':'Bob','Age':25,'Gender':'M','Score':78},{'Name':'Charlie','Age':22,'Gender':'M','Score':92},{'Name':'David','Age':24,'Gender':'M','Score':70},{'Name':'Eva','Age':23,'Gender':'F','Score':88}])
```

```python
df1=df.head(3)
df1
```

```python
df[df['Gender']=='F']
```

```python
df['Score1']=df['Score']
df['Score2']=df['Score']
df
```

```python
df['Age1']=df['Age']
df
```

```python
df=df.drop('Age1',axis=1)
df
```

```python
df=df.drop(['Score1','Score2'],axis=1)
df
```

```python
df=df.drop(3)
df
```

```python
df=(df[(df['Gender']=='F')&(df['Score']>85)])
df
```

```python
df=df.sort_values(by='Score',ascending=[False])
df
```

```python
def fillgrade(s):
    if(s>=90):
        return 'A'
    elif(s>=80 and s<90):
        return 'B'
    elif(s>=70 and s<80):
        return 'C'
    else:
        return 'D'
df['Grade']=df['Score'].apply(fillgrade)
df
```

```python
df1=pd.DataFrame([{'Month':'Jan','Product':'A','Sales':150},{'Month':'Feb','Product':'A','Sales':200},{'Month':'Mar','Product':'A','Sales':250},{'Month':'Jan','Product':'B','Sales':300},{'Month':'Feb','Product':'B','Sales':400},{'Month':'Mar','Product':'B','Sales':500}])
df1
```

```python
df1=df1.groupby('Product')['Sales'].sum()
df1
```

```python
df1=df1.loc[df1.groupby('Product')['Month'].idxmax()]
df1['Month']
```

```python
df1['Sales_Percentage']=(df1['Sales']/df1['Sales'].sum())*100
df1
```

```python
df2=pd.DataFrame([{'Name':'John','Department':'HR','Salary':50000},{'Name':'Doe','Department':'Finance','Salary':60000},{'Name':'Jane','Department':None,'Salary':55000},{'Name':'Anna','Department':'IT','Salary':None},{'Name':'Smith','Department':'HR','Salary':58000}])
df2
```
